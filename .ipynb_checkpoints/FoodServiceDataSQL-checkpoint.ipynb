```python
import pandas as pd
import zipfile as z
import sqlite3
import os

os.getcwd()
```

```python
zf = z.ZipFile("Data/FoodServiceData2020.zip","r")
zf.extractall("Data")
zf.close()
```

```python
# create the sqlite DB
db = sqlite3.connect("Data/MyDemo.db")

# read CSV into dataframe
df1 = pd.read_csv("Data/FoodServiceData2020.csv", header=None)

# convert the pandas dataframe to a sqlite table
df1.to_sql("foodscores", db, if_exists="replace")

def run_query(query):
    return pd.read_sql_query(query,db)
```

```python
# using os, we delete the big csv, if exists
target_file = "Data/FoodServiceData2020.csv"
  
if os.path.isfile(target_file): 
    os.remove(target_file) 
    print("File deleted") 
else: 
    print(f'"{target_file}" not found')
```

```python
query = """SELECT * FROM sqlite_master ORDER BY name;"""
run_query(query)
```

```python
#  Show entire table
query = """SELECT * FROM foodscores;"""
run_query(query)
```

```python
#  Displaying Organised Dataset with specific columns selected and renamed

#  NOTE still trying to figure out how to eliminate Row 0
query = """
SELECT "15" AS "Business Name", "13" AS "Score", "14" AS "Grade", "4" AS "Street", "6" AS "City",
"7" AS "State", "8" AS "ZIP", "10" AS "Latitude", "11" AS "Longitude"
FROM foodscores;
"""
run_query(query)
```

```python
# Rerunning above query with WHERE clause to filter out A ratings
query = """
SELECT "15" AS "Business Name", "13" AS "Score", "14" AS "Grade", "4" AS "Street", "6" AS "City",
"7" AS "State", "8" AS "ZIP", "10" AS "Latitude", "11" AS "Longitude"
FROM foodscores WHERE "14" = "A";
"""
run_query(query)
```

```python
#  Added a filter to target restaurant names e.g. "donald" for McDonalds
text = "donald"

query = """
SELECT 
"15" AS "Business Name", "13" AS "Score", "14" AS "Grade", "4" AS "Street", "6" AS "City", 
"7" AS "State", "8" AS "ZIP", "10" AS "Latitude", "11" AS "Longitude"
FROM foodscores WHERE "15" LIKE "%{}%";
""".format(text)
run_query(query)
```
