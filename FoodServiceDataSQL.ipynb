```python
import pandas as pd
import zipfile as z
import sqlite3
import os

os.getcwd()
```

```python
zf = z.ZipFile("Data/FoodServiceData2020.zip","r")
zf.extractall("Data")
zf.close()
```

```python
# create the sqlite DB
db = sqlite3.connect("Data/Ratings.db")

# read CSV into dataframe
df1 = pd.read_csv("Data/FoodServiceData2020.csv", header=0)

# convert the pandas dataframe to a sqlite table
df1.to_sql("foodscores", db, if_exists="replace")

def run_query(query):
    return pd.read_sql_query(query,db)
```

```python
# using os, we delete the big csv, if exists
target_file = "Data/FoodServiceData2020.csv"
  
if os.path.isfile(target_file): 
    os.remove(target_file) 
    print("File deleted") 
else: 
    print(f'"{target_file}" not found')
```

```python
query = """SELECT * FROM sqlite_master ORDER BY name;"""

run_query(query)
```

```python
# get information about my staging table 
query_table_help = """
SELECT name , *FROM PRAGMA_TABLE_INFO('foodscores');
"""
run_query(query_table_help)
```

```python
#  Displaying Organised Dataset with specific columns selected and renamed

query = """
SELECT "EstablishmentID" AS "ID", "EstablishmentName" AS "Business Name",
"Score", "Grade", "Address", "City", "State", "ZIP", "Latitude", "Longitude"
FROM foodscores;
"""

run_query(query)
```

```python
#  Rerunning above query with WHERE clause to filter out specified ratings

grade = "A"

query = """
SELECT "EstablishmentID" AS "ID", "EstablishmentName" AS "Business Name",
"Score", "Grade", "Address", "City", "State", "ZIP", "Latitude", "Longitude"
FROM foodscores WHERE "Grade" = "{}";
""".format(grade)

# run_query(query)
```

```python
#  Added a filter to target restaurant names e.g. "donald" for McDonalds

#  Search variable.  Change with all or part of restaurant name to filter.
text = "donald"

query = """
SELECT 
"EstablishmentID" AS "ID", "EstablishmentName" AS "Business Name",
"Score", "Grade", "Address", "City", "State", "ZIP", "Latitude", "Longitude"
FROM foodscores WHERE "Business Name" LIKE "%{}%";
""".format(text)

# run_query(query)
```

```python
#  Running a search for duplicate ratings
#  There are duplicate ratings but this is for a whole year so several restaurants got graded many times

query = """
SELECT "EstablishmentID" AS "ID", "EstablishmentName" AS "Business Name", COUNT(*) AS "Inspections"
FROM foodscores GROUP BY "ID", "Business Name" HAVING COUNT(*) > 1
ORDER BY COUNT(*) DESC LIMIT 10;"""

#  run_query(query)
```

```python
# Counting the number of ratings

query = """
SELECT "Grade", COUNT("Grade") AS "Count"
FROM foodscores GROUP BY "Grade" ORDER BY COUNT("Grade") DESC;
"""

# run_query(query)
```

```python
# drop old table if exists 
c = db.cursor()

Drop_TableName = "food_scores_cleaned"

Drop_SQL = f"""
drop table if exists {Drop_TableName};
"""

c.execute(Drop_SQL)

c.close()
```

```python
# create new table if not exists 
c = db.cursor()

Create_TableName = "food_scores_cleaned"

Create_SQL = f"""
create table if not exists {Create_TableName} AS
SELECT "EstablishmentID" AS "ID", "EstablishmentName" AS "Business Name",
"Score", "Grade", "Address", "City", "State", "ZIP", "Latitude", "Longitude"
FROM foodscores;
"""

c.execute(Create_SQL)

c.close()

clean_query = """
SELECT * FROM food_scores_cleaned
ORDER BY "ID";
"""

run_query(clean_query)
```
